### **Langkah (Step) 1: Import beberapa library yang akan digunakan adalah numpy, pandas, cv2, skimage, PIL, matplotlib.**




````python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io, transform
from PIL import Image
import matplotlib.pylab as plt
````

### **Langkah (Step) 2: Membaca Image dari URLs.**

````python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://assets.rebelmouse.io/eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJpbWFnZSI6Imh0dHBzOi8vYXNzZXRzLnJibC5tcy80MTQwNzA5L29yaWdpbi5qcGciLCJleHBpcmVzX2F0IjoxNzI4MDQ5MjAyfQ.HCk1uahPNm1ne0lNFZxAABdwt_kuuKHus9ZBXDlsqRA/img.jpg?width=600&quality=85"]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
````

````python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
````

### **Langkah (Step) 4: Mengakses pixel dengan memberi garis diagonal menyilang.**

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi) / 2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

# **Tugas Kelompok**
Saya mencoba untuk menutupi bagian TTL dan Jenis Kelamin dari kartu identitas berikut.

````python
import cv2 as cv
from skimage import io
import requests
import numpy as np
from PIL import Image
import io as io_pil
import matplotlib.pyplot as plt

# Download image from URL
image_url = "https://about.lovia.id/wp-content/uploads/2020/05/ktp-768x495.jpg"
response = requests.get(image_url)

# Convert image from bytes to a NumPy array
image_data = np.array(Image.open(io_pil.BytesIO(response.content)))

# Convert the image to OpenCV format (RGB to BGR)
image_rgb1 = cv.cvtColor(image_data, cv.COLOR_RGB2BGR)
image_rgb2 = cv.cvtColor(image_data, cv.COLOR_RGB2BGR)

# Menutupi bagian TTL (Tempat Tanggal Lahir)
start_x_ttl = 195  # X mulai dari TTL
start_y_ttl = 165  # Y mulai dari TTL
width_box_ttl = 210  # Lebar kotak penutup TTL
height_box_ttl = 17  # Tinggi kotak penutup TTL

for i in range(height_box_ttl):
    for j in range(width_box_ttl):
        image_rgb2[start_y_ttl + i, start_x_ttl + j] = [255, 0, 0]  # Tutup TTL dengan warna merah

# Menutupi bagian Jenis Kelamin
start_x_jk = 195  # X mulai dari Jenis Kelamin
start_y_jk = 190  # Y mulai dari Jenis Kelamin
width_box_jk = 120  # Lebar kotak penutup Jenis Kelamin
height_box_jk = 15  # Tinggi kotak penutup Jenis Kelamin

for i in range(height_box_jk):
    for j in range(width_box_jk):
        image_rgb2[start_y_jk + i, start_x_jk + j] = [0, 255, 255]  # Tutup Jenis Kelamin dengan warna kuning

# Gabungkan gambar asli dan gambar yang telah diedit
combined_frame = cv.hconcat((image_rgb1, image_rgb2))

# Convert back to RGB for matplotlib
combined_frame_rgb = cv.cvtColor(combined_frame, cv.COLOR_BGR2RGB)

# Tampilkan gambar menggunakan matplotlib
plt.imshow(combined_frame_rgb)
plt.axis('off')  # Hide axes
plt.show()
````
